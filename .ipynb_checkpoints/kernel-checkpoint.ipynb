```python
def run_query(q):
    with sqlite3.connect('database.sqlite') as conn:
        return pd.read_sql(q, conn)
```

```python
def run_command(c):
    with sqlite3.connect('database.sqlite') as conn:
        conn.isolation_level = None
        conn.execute(c)
```

```python
def show_tables():
    query= '''
    SELECT
        name,
        type
    FROM sqlite_master
    WHERE type IN ("table","view");
    ''' 
    return run_query(query)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Creating a bar plot
sns.barplot(x=nr, y=nr.index)
# Add labels to your graph
plt.xlabel('Count')
plt.ylabel('Negative Reason')
plt.title("Visualizing Most Common Negative Review Reasons")
plt.legend()
plt.show()
```

```python
query = '''

SELECT airline, 
       Positive,
       Negative,
       Neutral,
       Total,
       Negative * 100 /(Total) AS Per_Negt,       
       Neutral*100 / (Total) AS Per_Neu
       
FROM (
SELECT airline,
       SUM(CASE WHEN airline_sentiment = 'positive' THEN 1 ELSE 0 END) AS Positive, 
       SUM(CASE WHEN airline_sentiment = 'negative' THEN 1 ELSE 0 END) AS Negative,
       SUM(CASE WHEN airline_sentiment = 'neutral' THEN 1 ELSE 0 END) AS Neutral,
       COUNT(*) AS Total
FROM tweets
GROUP BY airline
)
ORDER BY Per_Negt DESC
'''
tweets = run_query(query)
tweets
```

```python
# Negative Reviews
tweets2 = tweets[["airline","Per_Negt"]]
```

```python
# Creating a bar plot
sns.barplot(y=tweets2["Per_Negt"], x=tweets2["airline"])
# Add labels to your graph
plt.xlabel('Airline')
plt.ylabel('Percentage of Negative Reviews')
plt.title("Visualizing Percentages of Negative Tweets for Airlines")
plt.legend()
plt.show()
```

It seems like most reviews or tweets about airlines are typically negative reviews as all 6 American airlines have more negative reviews than positive reviews. However, there is a large discrepancy between the airline with the highest and lowest percentage of negative reviews of 35%. 


```python
# Neutral Reviews
tweets3 = tweets[["airline","Per_Neu"]]
```

```python
# Creating a bar plot
sns.barplot(y=tweets3["Per_Neu"], x=tweets3["airline"])
# Add labels to your graph
plt.xlabel('Airline')
plt.ylabel('Percentage of Neutral Reviews')
plt.title("Visualizing Percentages of Neutral Tweets for Airlines")
plt.legend()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
# print wordcloud
show_wordcloud(tweet["text"])

# command prompt - python -m pip install wordcloud
```

As we can see above, most of the words are related to flights and not about the flight experience.  AmericanAir, VirginAmerica, Flight, really, love Thank, call. It is difficult to judge by looking at individual words as most are nouns and neutral words. 
Positive: thank, love, great, amazing, 
Negative: ugh, suicide, bad, missed, worry
